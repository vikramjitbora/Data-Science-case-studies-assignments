## Data Cleaning

#### Missing Value Analysis

```python
# Defining a function to return the percentage of null values in dataframe and to return only columns which have null values
def null_perc(dataframe):
    '''
    This function will return the count of null values in dataframe and return only columns which have null values
    '''
    perc = dataframe.isna().sum()/len(dataframe) * 100
    perc.sort_values(ascending=False,inplace=True)
    return perc[perc>0]
```

Checking the percentage of null values in the data

```python
null_perc(df)
```

Columns with more than 45% data missing

```python
nulls = null_perc(df)
nulls_45A = nulls[nulls>45]
nulls_45A
```

```python
#Number of columns with >45% null values

len(nulls_45A)
```

We will be dropping columns with more than 45% missing values from the dataset before proceeding with our analysis

```python
cols = nulls_45A.index.to_list()
df.drop(columns=cols,inplace=True)
```

```python
df.shape
```

```python
## Checking remaining columns with null values
null_perc(df)
```

```python
## Removing rows where columns have less than 1% null values
df=df[~df.OBS_30_CNT_SOCIAL_CIRCLE.isnull()]
df=df[~df.DEF_30_CNT_SOCIAL_CIRCLE.isnull()]
df=df[~df.OBS_60_CNT_SOCIAL_CIRCLE.isnull()]
df=df[~df.DEF_60_CNT_SOCIAL_CIRCLE.isnull()]
df=df[~df.AMT_ANNUITY.isnull()]
df=df[~df.AMT_GOODS_PRICE.isnull()]
df=df[~df.NAME_TYPE_SUITE.isnull()]
df=df[~df.CNT_FAM_MEMBERS.isnull()]
df=df[~df.DAYS_LAST_PHONE_CHANGE.isnull()]
df=df[~df.EXT_SOURCE_2.isnull()]
```

```python
## Imputing missing values using Central Tendency
values = {
          'EXT_SOURCE_3' : df.EXT_SOURCE_3.median(),
          'OCCUPATION_TYPE' : df.OCCUPATION_TYPE.mode()[0],
          'AMT_REQ_CREDIT_BUREAU_HOUR' : df.AMT_REQ_CREDIT_BUREAU_HOUR.median(),
          'AMT_REQ_CREDIT_BUREAU_DAY' : df.AMT_REQ_CREDIT_BUREAU_DAY.median(),
          'AMT_REQ_CREDIT_BUREAU_WEEK' : df.AMT_REQ_CREDIT_BUREAU_WEEK.median(),
          'AMT_REQ_CREDIT_BUREAU_QRT' : df.AMT_REQ_CREDIT_BUREAU_QRT.median(),
          'AMT_REQ_CREDIT_BUREAU_MON' : df.AMT_REQ_CREDIT_BUREAU_MON.median(),
          'AMT_REQ_CREDIT_BUREAU_YEAR' : df.AMT_REQ_CREDIT_BUREAU_YEAR.median()
         }
```

```python
df.fillna(value=values,inplace=True)
```

```python
null_perc(df)
```

### Numerical variables

```python
def univariate(col):
    '''
    # Function for univariate analysis for numerical columns. 
    1st row will plot the distribution and box plot for the variable with respecct to target value 0. 
    The next row will plot the graphs for target variable 1
    '''

    fig,ax = plt.subplots(nrows=2,ncols=2, figsize=[15,15])

    sns.distplot(ax=ax[0][0],a=target0[col])
    ax[0][0].set_title('Customers without payment difficulties - Distribution plot \n',fontsize=12,fontweight = 'bold',color='brown')
    sns.boxplot(ax=ax[0][1],x=target0[col])
    ax[0][1].set_title('Customers without payment difficulties - Box Plot \n',fontsize=12,fontweight = 'bold',color='brown')

    sns.distplot(ax=ax[1][0],a=target1[col]) 
    ax[1][0].set_title('Customers with payment difficulties - Distribution plot \n',fontsize=12,fontweight = 'bold',color='brown')
    sns.boxplot(ax=ax[1][1],x=target1[col])
    ax[1][1].set_title('Customers with payment difficulties - Box Plot \n',fontsize=12,fontweight = 'bold',color='brown')

    plt.show()
```

### Correlation

We will find the correlation for both dataframes with target 0 and target 1

```python
correlation_cols = ['AMT_GOODS_PRICE','AMT_INCOME_TOTAL','AMT_ANNUITY','AMT_CREDIT','YEARS_EMPLOYED','YEARS_BIRTH',
         'YEARS_REGISTRATION', 'YEARS_ID_PUBLISH','EXT_SOURCE_2','EXT_SOURCE_3']
```

```python
corr_matrix0 = target0[correlation_cols].corr()
corr_matrix1 = target1[correlation_cols].corr()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2, figsize=[20,8])

sns.heatmap(ax=axes[0],data=corr_matrix0,annot=True,cmap='YlGnBu', fmt='.2f', linewidths=0.05,linecolor='grey')
axes[0].set_title('Correlation matrix for target variable 0 \n',fontsize=12,fontweight = 'bold',color='brown')

sns.heatmap(ax=axes[1],data=corr_matrix1,annot=True,cmap='YlGnBu',fmt='.2f',linewidths=0.05, linecolor='grey')
axes[1].set_title('Correlation matrix for target variable 1 \n', fontsize=12,fontweight = 'bold',color='brown')

plt.show()
```

Now we need to find out the top 10 correlation pairs from each of the dataframes

```python
#the matrix is symmetric so we need to extract upper triangle matrix without diagonal (k = 1)

top10_corr0 = (corr_matrix0.where(np.triu(np.ones(corr_matrix0.shape), k=1).astype(np.bool))
                  .stack()
                  .sort_values(ascending=False)).head(10)
top10_corr0
```

```python
top10_corr1 = (corr_matrix1.where(np.triu(np.ones(corr_matrix1.shape), k=1).astype(np.bool))
                  .stack()
                  .sort_values(ascending=False)).head(10)
top10_corr1
```

We observe that among the top 10 correlations from both the dataframes, the top 3 correlations for both are almost similar.


- AMT_GOODS_PRICE and AMT_CREDIT, 
- AMT_GOODS_PRICE and AMT_ANNUITY, 
- AMT_ANNUITY and AMT_CREDIT

Some observation for Missing value imputation:
    1. NAME_TYPE_SUITE  - 49% data is missing. So adding another category as 'Missing' to replace the missing values
    2. NAME_YIELD_GROUP - Middle, high and low_normal are all almost equally distributed. However, since almost 31% of the data is missing in this column, we will be addind another category as 'Missing' to replace the missing values
    3. PRODUCT_COMBINATION - Since a lot of product combinations have very similar distribution and there are relatively a very small percentage of data missing, we will not be imputing the missing values for this column for our analysis
    4. NAME_CONTRACT_TYPE - Since Cash Loans and Consumer Loans have similar distribution and there are relatively a very small percentage of data missing, we will not be imputing the missing values for this column for our analysis
    
For other columns, we are using Central Tendency to impute missing values.

```python
values = {'NAME_TYPE_SUITE' : 'Missing',
          'NAME_YIELD_GROUP' : 'Missing',
          'NFLAG_INSURED_ON_APPROVAL' : prev_app.NFLAG_INSURED_ON_APPROVAL.mode()[0],
          'NAME_PAYMENT_TYPE' : prev_app.NAME_PAYMENT_TYPE.mode()[0],
          'AMT_ANNUITY' : prev_app.AMT_ANNUITY.median(),
          'AMT_GOODS_PRICE' : prev_app.AMT_GOODS_PRICE.median(),
          'CNT_PAYMENT' : prev_app.CNT_PAYMENT.median(),
          'NAME_PORTFOLIO' : prev_app.NAME_PORTFOLIO.mode()[0],
          'NAME_CLIENT_TYPE' : prev_app.NAME_CLIENT_TYPE.mode()[0],
          'DAYS_LAST_DUE' : prev_app.DAYS_LAST_DUE.median(),
          'DAYS_LAST_DUE_1ST_VERSION' : prev_app.DAYS_LAST_DUE_1ST_VERSION.median(),
          'DAYS_FIRST_DUE' : prev_app.DAYS_FIRST_DUE.median(),
          'DAYS_FIRST_DRAWING' : prev_app.DAYS_FIRST_DRAWING.median(),
          'DAYS_TERMINATION' : prev_app.DAYS_TERMINATION.median()
         }
```

```python
prev_app.fillna(value=values,inplace=True)
```

```python
null_perc(prev_app)
```

```python
prev_app[prev_app.AMT_CREDIT.isna()]
```

```python
prev_app[prev_app.NAME_CONTRACT_TYPE.isna()]
```

Moreover, whenever one category has NaNs, the other category also has NaNs. As such, we can drop the rows with missing values for these categories.

```python
# A function to plot bar graph for different columns for each of the data frames

def multiple_plots(col):
    '''
    This function will plot the bar graph for different columns taken as a parameter with respect to Refused, approved, Cancelled and Unused offer 
    contract status of the loans
    '''
    
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,ncols=2, figsize=(15,12), sharey=True)
    fig.tight_layout(pad=10.0)

    ax1 = sns.countplot(ax=ax1,data=approved_df,x=col)
    ax1.set_title("Refused", fontsize=12,fontweight = 'bold',color='brown')
    ax1.set_ylabel("Number of Loans",fontsize=10)
    ax1.set_xticklabels(ax1.get_xticklabels(),rotation=45)

    ax2 = sns.countplot(ax=ax2,data=refused_df,x=col)
    ax2.set_title("Approved", fontsize=12,fontweight = 'bold',color='brown')
    ax2.set_ylabel("Number of Loans",fontsize=10 )
    ax2.set_xticklabels(ax2.get_xticklabels(),rotation=45)

    ax3 = sns.countplot(ax=ax3,data=canceled_df,x=col)
    ax3.set_title("Canceled", fontsize=12,fontweight = 'bold',color='brown')
    ax3.set_ylabel("Number of Loans",fontsize=10)
    ax3.set_xticklabels(ax3.get_xticklabels(),rotation=45)

    ax4 = sns.countplot(ax=ax4,data=unused_df,x=col)
    ax4.set_title("Unused", fontsize=12,fontweight = 'bold',color='brown')
    ax4.set_ylabel("Number of Loans",fontsize=10,color='brown')
    ax4.set_xticklabels(ax4.get_xticklabels(),rotation=45)
       
    plt.show()
```

- The count of Lower Secondary in NAME_EDUCATION_TYPE is comparatively less and it also has high % of payment difficulties. Hence, client with education type as Lower Secondary are the driving factors for Loan Defaulters.

- Male senior cooking staffs and male very young HR staffs are having maximum difficuly in loan payment. 

- Customers not having cars along with house/flat are having difficulty in making loan payments.

- Male customers who are not reachable on phone and have low incomes are having difficulty in making loan payments.

- Loan payment difficulties are seen more in more-membered families as compared to less-membered families.

- Customers with 'REGION_POPULATION_RELATIVE' < 0.03 are having more difficulties in paying for loans. Also, higher the customer's Region Rating, higher the chances of his/her defaulting.

- Customers who changed their registration in last 12 years and ID in 9 years before the application, are having more difficulties in paying for loans as compared to those who have changed before 12 and 9 years timeframe respectively.

- Non-senior citizens(less than 56 years age) who have spent less than 8.8 years in their current employment are having more difficulties in paying for loans.

- Customers who have received ratings lower than 0.57 and 0.53 from external sources 2 and 3 are having more difficulties in paying for loans
